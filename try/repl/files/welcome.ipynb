# BAM Engine Interactive Demo

Run BAM Engine in your browser via Pyodide.

```python
import micropip
await micropip.install('bamengine')
print('bamengine installed!')
```

```python
import bamengine as bam

sim = bam.Simulation.init(seed=42)
results = sim.run(n_periods=100, collect=True)

print(f'Periods: {sim.t}')
print(f'Unemployment: {sim.ec.unemp_rate_history[-1]:.2%}')
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(sim.ec.unemp_rate_history)
ax1.set_title('Unemployment Rate')
ax1.set_xlabel('Period')
ax1.grid(True, alpha=0.3)

ax2.plot(sim.ec.avg_mkt_price_history)
ax2.set_title('Average Market Price')
ax2.set_xlabel('Period')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
